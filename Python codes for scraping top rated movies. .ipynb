```python
!pip install html5lib
import requests
from bs4 import BeautifulSoup
import pandas as pd
import csv

imdb_url = 'https://www.imdb.com/list/ls009997493/'
page = requests.get(imdb_url)
soup = BeautifulSoup(page.content, 'html.parser')
movie_title = soup.find_all('h3', class_='lister-item-header')
movie_rating = soup.find_all('div', class_='ipl-rating-star small')

movie_titles = []
movie_ratings = []

for i in movie_title:
    movie_titles.append(i.text.replace('\n', ''))

for i in movie_rating:
    rating = i.find('span', class_='ipl-rating-star__rating')
    movie_ratings.append(rating.text if rating else 'N/A')
# Create a list of tuples containing the movie titles and ratings
movies_list = list(zip(movie_titles, movie_ratings))
# Save the list as a CSV file
with open('movies.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Title', 'Rating'])
    for movie in movies_list:
        writer.writerow(movie)

# Print the list of movies and ratings
for movie in movies_list:
    print(f"{movie[0]} - {movie[1]}")
```
